```python
import pandas as pd
import os

# 1. Setup the directory structure
# Note: 'mfn ave' and 'pref ave' contain spaces as per your last message
base_path = "kazakhstan"
folders = {
    "MFN": "mfn",
    "MFN_AVE": "mfn ave",
    "PREF": "pref",
    "PREF_AVE": "pref ave"
}

# 2. Define standard WITS file prefixes for Kazakhstan
# Standard MFN: MFN_KAZ_... | AVE MFN: AVEMFN_KAZ_... 
# Standard Pref: Pref_KAZ_... | AVE Pref: AVEPref_KAZ_...
prefixes = {
    "MFN": "MFN_KAZ_",
    "MFN_AVE": "AVEMFN_KAZ_",
    "PREF": "Pref_KAZ_",
    "PREF_AVE": "AVEPref_KAZ_"
}

years = range(2013, 2024)

def merge_wits_files(category):
    all_data = []
    folder = folders[category]
    prefix = prefixes[category]
    
    print(f"--- Merging {category} ---")
    
    for year in years:
        file_path = f"{base_path}/{folder}/{prefix}{year}.CSV"
        
        if os.path.exists(file_path):
            try:
                # low_memory=False ensures that mixed data types (common in WITS) 
                # don't trigger warnings or errors.
                df = pd.read_csv(file_path, low_memory=False)
                
                # Critical step for Teti/Fontagne: Strip hidden spaces from headers
                # so that columns like 'ProductCode ' become 'ProductCode'
                df.columns = df.columns.str.strip()
                
                all_data.append(df)
                print(f"Added: {year}")
            except Exception as e:
                print(f"Error reading {year}: {e}")
        else:
            print(f"Missing: {file_path}")
            
    if all_data:
        # sort=False ensures we keep the original column order from the files
        merged_df = pd.concat(all_data, axis=0, ignore_index=True, sort=False)
        print(f"Final {category} dataset shape: {merged_df.shape}")
        return merged_df
    else:
        print(f"No data found for {category}")
        return pd.DataFrame()

# 3. Execute the merge
df_mfn = merge_wits_files("MFN")
df_mfn_ave = merge_wits_files("MFN_AVE")
df_pref = merge_wits_files("PREF")
df_pref_ave = merge_wits_files("PREF_AVE")

# 4. Check if ProductCode needs cleaning (sometimes exported with ' marks)
def clean_product_codes(df):
    if not df.empty and 'ProductCode' in df.columns:
        df['ProductCode'] = df['ProductCode'].astype(str).str.replace("'", "").str.zfill(6)
    return df

df_mfn = clean_product_codes(df_mfn)
df_mfn_ave = clean_product_codes(df_mfn_ave)
df_pref = clean_product_codes(df_pref)
df_pref_ave = clean_product_codes(df_pref_ave)
```

```python
import pandas as pd
import numpy as np

# --- 1. Load Reference Files ---
# Country List (Semicolon delimited)
cl = pd.read_csv('country_list.csv', sep=';', encoding='latin1')
cl.columns = cl.columns.str.strip()

# Preference Beneficiaries (Comma delimited)
pb = pd.read_csv('preference_beneficials.csv', encoding='latin1')
pb.columns = pb.columns.str.strip()

# --- 2. Build the Official ISO Mapping Dictionary ---
# Teti's correction map for non-standard WITS codes
iso_correction = {
    'ROM': 'ROU', 'ZAR': 'COD', 'TMP': 'TLS', 'SUD': 'SDN',
    'SAM': 'WSM', 'SER': 'SRB', 'MNE': 'MNE', 'ANT': 'CUW',
    'WLF': 'WLF', 'SVK': 'SVK'
}

# Create a mapping from Numerical Code (string padded) to Official ISO3
cl['CountryCode_str'] = cl['CountryCode'].astype(str).str.zfill(3)
# First map Numerical -> WITS ISO3
wits_iso_dict = cl.set_index('CountryCode_str')['ISO3'].to_dict()

def to_official_iso(num_code):
    num_str = str(int(num_code)).zfill(3)
    wits_iso = wits_iso_dict.get(num_str)
    # Apply Teti corrections if applicable
    return iso_correction.get(wits_iso, wits_iso)

# --- 3. MFN ISO Mapping ---
print("Mapping MFN reporters to Official ISO3...")
df_mfn_combined['iso3_reporter'] = df_mfn_combined['Reporter_ISO_N'].apply(to_official_iso)

# --- 4. Preferential Bilateral Disaggregation ---
print("Disaggregating preferential schedules to individual exporters...")

# Merge dataset with beneficiaries
# RegionCode in pb matches 'Partner' (the schedule) in your dataset
# 'Partner' in pb is the numerical code of the beneficiary country
df_pref_bilateral = pd.merge(
    df_pref_combined,
    pb[['RegionCode', 'Partner']], 
    left_on='Partner', 
    right_on='RegionCode',
    how='inner',
    suffixes=('', '_beneficiary')
)

# Assign Official ISO codes to both sides
df_pref_bilateral['iso3_reporter'] = df_pref_bilateral['Reporter_ISO_N'].apply(to_official_iso)
df_pref_bilateral['iso3_exporter'] = df_pref_bilateral['Partner_beneficiary'].apply(to_official_iso)

# --- 5. Teti's Aggregation and Filtering ---
print("Applying Teti's 'Lowest Rate' and 'No-Increase' rules...")

# Rule 1: Multiple Agreements -> Lowest available rate
df_pref_bilateral = df_pref_bilateral.groupby(
    ['iso3_reporter', 'iso3_exporter', 'ProductCode', 'Year']
).agg({'pref': 'min', 'prefAVE': 'min'}).reset_index()

# Rule 2: Spike Filter (Treat increases over time as missing data)
df_pref_bilateral = df_pref_bilateral.sort_values(['iso3_reporter', 'iso3_exporter', 'ProductCode', 'Year'])

def flag_spikes(group):
    # Calculate tariff difference; if > 0, it's a 'false' spike
    group['diff'] = group['prefAVE'].diff()
    group.loc[group['diff'] > 0, ['pref', 'prefAVE']] = np.nan
    return group

df_pref_bilateral = df_pref_bilateral.groupby(
    ['iso3_reporter', 'iso3_exporter', 'ProductCode'], 
    group_keys=False
).apply(flag_spikes)

# Drop helper columns
df_pref_bilateral = df_pref_bilateral.drop(columns=['diff'])

print("\n--- Phase 1.8 Success ---")
print(f"Bilateral Pairs generated: {df_pref_bilateral['iso3_exporter'].nunique()} exporters found for Kazakhstan.")
display(df_pref_bilateral.head())
```

```python
import pandas as pd
import numpy as np
import os

# --- 1. BULLETPROOF MAPPING SETUP ---
print("Loading reference files...")
cl = pd.read_csv('country_list.csv', sep=';', encoding='latin-1', header=0)
cl.columns = ['ISO3', 'CountryCode', 'CountryName']

pb = pd.read_csv('preference_beneficials.csv', sep=',', encoding='latin-1', header=0)
pb.columns = ['RegionCode', 'RegionName', 'Partner', 'PartnerName']

iso_correction = {
    'ROM': 'ROU', 'ZAR': 'COD', 'TMP': 'TLS', 'SUD': 'SDN',
    'SAM': 'WSM', 'SER': 'SRB', 'MNE': 'MNE', 'ANT': 'CUW',
    'WLF': 'WLF', 'SVK': 'SVK'
}

cl['CountryCode'] = pd.to_numeric(cl['CountryCode'], errors='coerce')
iso_dict = cl.set_index('CountryCode')['ISO3'].to_dict()

def to_official_iso(numeric_code):
    try:
        wits_iso = iso_dict.get(int(float(numeric_code)))
        return iso_correction.get(wits_iso, wits_iso)
    except:
        return np.nan

pb['iso3_exporter'] = pb['Partner'].apply(to_official_iso)

# --- 2. PREPARE THE LOOPS ---
mfn_drop = ['Sum_Of_Rates', 'Min_Rate', 'Max_Rate', 'Nbr_NA_Lines', 'Nbr_Free_Lines', 
            'Nbr_AVE_Lines', 'Nbr_Dutiable_Lines', 'TotalNoOfValidLines', 'TotalNoOfLines', 'NomenCode']

pref_drop = ['ExcludedFrom', 'Sum_Of_Rates', 'Min_Rate', 'Max_Rate', 'TotalNoOfLines', 
             'Nbr_Pref_Lines', 'Nbr_MFN_Lines', 'Nbr_NA_Lines', 'NomenCode', 'EstCode']

def clean_tariff_file(file_path, rate_name, drop_list):
    if not os.path.exists(file_path):
        print(f"  -> Missing file: {file_path}")
        return pd.DataFrame()
    
    df = pd.read_csv(file_path, encoding='latin-1')
    if 'SimpleAverage' in df.columns:
        df = df.rename(columns={'SimpleAverage': rate_name})
    df = df.drop(columns=[c for c in drop_list if c in df.columns], errors='ignore')
    if 'ProductCode' in df.columns:
        df['ProductCode'] = df['ProductCode'].astype(str).str.replace("'", "").str.zfill(6)
    return df

# Storage for the yearly dataframes
all_mfn_years = []
all_pref_years = []

# --- 3. PROCESS EVERY YEAR (2013-2023) ---
print("Processing yearly files...")
for year in range(2013, 2024):
    print(f"Processing Year {year}...")
    
    # Construct exact paths using your provided structure
    mfn_path = f"kazakhstan/mfn/MFN_KAZ_{year}.CSV"
    mfn_ave_path = f"kazakhstan/mfn ave/AVEMFN_KAZ_{year}.CSV"
    pref_path = f"kazakhstan/pref/Pref_KAZ_{year}.CSV"
    pref_ave_path = f"kazakhstan/pref ave/AVEPref_KAZ_{year}.CSV"
    
    # Load and clean
    df_mfn = clean_tariff_file(mfn_path, 'mfn', mfn_drop)
    df_mfn_ave = clean_tariff_file(mfn_ave_path, 'mfnAVE', mfn_drop)
    df_pref = clean_tariff_file(pref_path, 'pref', pref_drop)
    df_pref_ave = clean_tariff_file(pref_ave_path, 'prefAVE', pref_drop)
    
    # Merge MFN for this year
    if not df_mfn.empty and not df_mfn_ave.empty:
        df_mfn_merged = pd.merge(df_mfn, df_mfn_ave, on=['Reporter_ISO_N', 'Year', 'ProductCode'], how='outer')
        all_mfn_years.append(df_mfn_merged)
        
    # Merge and Disaggregate PREF for this year
    if not df_pref.empty and not df_pref_ave.empty:
        df_pref_merged = pd.merge(df_pref, df_pref_ave, on=['Reporter_ISO_N', 'Year', 'ProductCode', 'Partner'], how='outer')
        
        # Explode to bilateral pairs using the pre-mapped beneficiaries list
        df_pref_bilateral = pd.merge(
            df_pref_merged, 
            pb[['RegionCode', 'iso3_exporter']], 
            left_on='Partner', 
            right_on='RegionCode', 
            how='inner'
        )
        all_pref_years.append(df_pref_bilateral)

# --- 4. CONCATENATE ALL YEARS INTO MASTER PANELS ---
print("\nConcatenating 2013-2023 panels...")
master_mfn = pd.concat(all_mfn_years, ignore_index=True)
master_pref = pd.concat(all_pref_years, ignore_index=True)

# Map Reporter ISOs
master_mfn['iso3_reporter'] = master_mfn['Reporter_ISO_N'].apply(to_official_iso)
master_pref['iso3_reporter'] = master_pref['Reporter_ISO_N'].apply(to_official_iso)

# --- 5. APPLY TETI TIME-SERIES RULES ---
print("Applying Teti (2024) Appendix F.2 rules across the panel...")

# Rule 1: Min Rate for overlapping agreements in the same year
master_pref = master_pref.groupby(
    ['iso3_reporter', 'iso3_exporter', 'ProductCode', 'Year']
).agg({'pref': 'min', 'prefAVE': 'min'}).reset_index()

# Rule 2: Preferential Spike Filter (Time-Series Check)
# Sort chronologically so .diff() checks the change from year to year
master_pref = master_pref.sort_values(['iso3_reporter', 'iso3_exporter', 'ProductCode', 'Year'])

def flag_spikes(group):
    # If the preferential rate goes UP from one year to the next, it's an error. Mark as NaN.
    group['diff'] = group['prefAVE'].diff()
    group.loc[group['diff'] > 0, ['pref', 'prefAVE']] = np.nan
    return group

master_pref = master_pref.groupby(
    ['iso3_reporter', 'iso3_exporter', 'ProductCode'], 
    group_keys=False
).apply(flag_spikes)

master_pref = master_pref.drop(columns=['diff'], errors='ignore')

# Save the final Master Panels!
master_mfn.to_csv('KAZ_MFN_Master_2013_2023.csv', index=False)
master_pref.to_csv('KAZ_PREF_Master_2013_2023.csv', index=False)

print("\n--- PHASE 1 COMPLETE! ---")
print(f"Master MFN Rows: {len(master_mfn)}")
print(f"Master PREF Rows: {len(master_pref)}")
```

```python
print("--- STEP 2: LOADING AND CLEANING CHINA DATA ---")

mfn_drop = ['Sum_Of_Rates', 'Min_Rate', 'Max_Rate', 'Nbr_NA_Lines', 'Nbr_Free_Lines', 
            'Nbr_AVE_Lines', 'Nbr_Dutiable_Lines', 'TotalNoOfValidLines', 'TotalNoOfLines', 'NomenCode']

pref_drop = ['ExcludedFrom', 'Sum_Of_Rates', 'Min_Rate', 'Max_Rate', 'TotalNoOfLines', 
             'Nbr_Pref_Lines', 'Nbr_MFN_Lines', 'Nbr_NA_Lines', 'NomenCode', 'EstCode']

def clean_tariff_file(file_path, rate_name, drop_list):
    if not os.path.exists(file_path):
        print(f"  [!] Missing file: {file_path}")
        return pd.DataFrame()
    
    df = pd.read_csv(file_path, encoding='latin-1')
    
    # BRUTE FORCE CLEANING: Strip spaces, double quotes, and single quotes from column names
    df.columns = [str(c).strip().replace('\ufeff', '').replace('"', '').replace("'", "") for c in df.columns]
    
    if 'SimpleAverage' in df.columns:
        df = df.rename(columns={'SimpleAverage': rate_name})
        
    df = df.drop(columns=[c for c in drop_list if c in df.columns], errors='ignore')
    
    if 'ProductCode' in df.columns:
        # Also clean the actual ProductCode values inside the data, just in case they have quotes too!
        df['ProductCode'] = df['ProductCode'].astype(str).str.replace("'", "").str.replace('"', '').str.strip().str.zfill(6)
        
    return df

all_mfn_years = []
all_pref_years = []

for year in range(2013, 2024):
    print(f" -> Processing {year}...")
    
    # The paths exactly as you specified
    mfn_path = f"china mfn/MFN_CHN_{year}.CSV"
    mfn_ave_path = f"china mfn ave/AVEMFN_CHN_{year}.CSV"
    pref_path = f"china /china pref/PREF_CHN_{year}.CSV"
    pref_ave_path = f"china /china pref ave/AVEPREF_CHN_{year}.CSV"
    
    # Load and clean
    df_mfn = clean_tariff_file(mfn_path, 'mfn', mfn_drop)
    df_mfn_ave = clean_tariff_file(mfn_ave_path, 'mfnAVE', mfn_drop)
    df_pref = clean_tariff_file(pref_path, 'pref', pref_drop)
    df_pref_ave = clean_tariff_file(pref_ave_path, 'prefAVE', pref_drop)
    
    # Merge MFN
    if not df_mfn.empty and not df_mfn_ave.empty:
        df_mfn_merged = pd.merge(df_mfn, df_mfn_ave, on=['Reporter_ISO_N', 'Year', 'ProductCode'], how='outer')
        all_mfn_years.append(df_mfn_merged)
        
    # Merge and Disaggregate PREF
    if not df_pref.empty and not df_pref_ave.empty:
        df_pref_merged = pd.merge(df_pref, df_pref_ave, on=['Reporter_ISO_N', 'Year', 'ProductCode', 'Partner'], how='outer')
        df_pref_bilateral = pd.merge(
            df_pref_merged, pb[['RegionCode', 'iso3_exporter']], 
            left_on='Partner', right_on='RegionCode', how='inner'
        )
        all_pref_years.append(df_pref_bilateral)

# Concatenate all years together
master_mfn = pd.concat(all_mfn_years, ignore_index=True)
master_pref = pd.concat(all_pref_years, ignore_index=True)

master_mfn['iso3_reporter'] = 'CHN'
master_pref['iso3_reporter'] = 'CHN'

print("\n--- STEP 2 SUCCESS! ---")
print("MFN Panel Columns:", master_mfn.columns.tolist())
print("PREF Panel Columns:", master_pref.columns.tolist())
print(f"Total MFN Rows: {len(master_mfn)}")
print(f"Total PREF Rows: {len(master_pref)}")
```

```python
import os
import pandas as pd
import numpy as np

print("--- STARTING ULTIMATE CHINA MASTER SCRIPT ---")

# ==========================================
# STEP 1: REFERENCE MAPS
# ==========================================
print("1. Loading reference maps...")
cl = pd.read_csv('country_list.csv', sep=';', encoding='latin-1')
cl.columns = [str(c).strip().replace('\ufeff', '').replace('"', '').replace("'", "") for c in cl.columns]
cl.columns = ['ISO3', 'CountryCode', 'CountryName']

pb = pd.read_csv('preference_beneficials.csv', sep=',', encoding='latin-1')
pb.columns = [str(c).strip().replace('\ufeff', '').replace('"', '').replace("'", "") for c in pb.columns]
pb.columns = ['RegionCode', 'RegionName', 'Partner', 'PartnerName']

iso_correction = {
    'ROM': 'ROU', 'ZAR': 'COD', 'TMP': 'TLS', 'SUD': 'SDN',
    'SAM': 'WSM', 'SER': 'SRB', 'MNE': 'MNE', 'ANT': 'CUW',
    'WLF': 'WLF', 'SVK': 'SVK'
}

cl['CountryCode'] = pd.to_numeric(cl['CountryCode'], errors='coerce')
iso_dict = cl.set_index('CountryCode')['ISO3'].to_dict()

def to_official_iso(numeric_code):
    try:
        wits_iso = iso_dict.get(int(float(numeric_code)))
        return iso_correction.get(wits_iso, wits_iso)
    except:
        return np.nan

pb['iso3_exporter'] = pb['Partner'].apply(to_official_iso)

# ==========================================
# STEP 2: LOAD AND MERGE TARIFFS
# ==========================================
print("2. Loading and cleaning yearly data...")
mfn_drop = ['Sum_Of_Rates', 'Min_Rate', 'Max_Rate', 'Nbr_NA_Lines', 'Nbr_Free_Lines', 
            'Nbr_AVE_Lines', 'Nbr_Dutiable_Lines', 'TotalNoOfValidLines', 'TotalNoOfLines', 'NomenCode']

pref_drop = ['ExcludedFrom', 'Sum_Of_Rates', 'Min_Rate', 'Max_Rate', 'TotalNoOfLines', 
             'Nbr_Pref_Lines', 'Nbr_MFN_Lines', 'Nbr_NA_Lines', 'NomenCode', 'EstCode']

def clean_tariff_file(file_path, rate_name, drop_list):
    if not os.path.exists(file_path):
        return pd.DataFrame()
    
    df = pd.read_csv(file_path, encoding='latin-1', low_memory=False)
    
    if 'SimpleAverage' in df.columns:
        df = df.rename(columns={'SimpleAverage': rate_name})
        
    df = df.drop(columns=[c for c in drop_list if c in df.columns], errors='ignore')
    
    if 'ProductCode' in df.columns:
        df['ProductCode'] = df['ProductCode'].astype(str).str.replace(".0", "", regex=False).str.zfill(6)
        
    return df

all_mfn, all_pref = [], []

for year in range(2013, 2024):
    mfn_path = f"china mfn/MFN_CHN_{year}.CSV"
    mfn_ave_path = f"china mfn ave/AVEMFN_CHN_{year}.CSV"
    pref_path = f"china /china pref/PREF_CHN_{year}.CSV"
    pref_ave_path = f"china /china pref ave/AVEPREF_CHN_{year}.CSV"
    
    m = clean_tariff_file(mfn_path, 'mfn', mfn_drop)
    ma = clean_tariff_file(mfn_ave_path, 'mfnAVE', mfn_drop)
    p = clean_tariff_file(pref_path, 'pref', pref_drop)
    pa = clean_tariff_file(pref_ave_path, 'prefAVE', pref_drop)
    
    if not m.empty and not ma.empty:
        all_mfn.append(pd.merge(m, ma, on=['Reporter_ISO_N', 'Year', 'ProductCode'], how='outer'))
        
    if not p.empty and not pa.empty:
        pref_merge = pd.merge(p, pa, on=['Reporter_ISO_N', 'Year', 'ProductCode', 'Partner'], how='outer')
        pref_bilateral = pd.merge(pref_merge, pb[['RegionCode', 'iso3_exporter']], left_on='Partner', right_on='RegionCode', how='inner')
        all_pref.append(pref_bilateral)

master_mfn = pd.concat(all_mfn, ignore_index=True)
master_pref = pd.concat(all_pref, ignore_index=True)

master_mfn['iso3_reporter'] = 'CHN'
master_pref['iso3_reporter'] = 'CHN'

# ==========================================
# STEP 3: TETI SPIKE FILTER (VECTORIZED)
# ==========================================
print("3. Applying time-series spike filter...")
master_pref = master_pref.groupby(
    ['iso3_reporter', 'iso3_exporter', 'ProductCode', 'Year']
).agg({'pref': 'min', 'prefAVE': 'min'}).reset_index()

master_pref = master_pref.sort_values(['iso3_reporter', 'iso3_exporter', 'ProductCode', 'Year'])

# Safe Vectorized calculation!
master_pref['diff'] = master_pref.groupby(['iso3_reporter', 'iso3_exporter', 'ProductCode'])['prefAVE'].diff()
master_pref.loc[master_pref['diff'] > 0, ['pref', 'prefAVE']] = np.nan
master_pref = master_pref.drop(columns=['diff'])

# ==========================================
# STEP 4: SKELETON MATRIX & INTERPOLATION
# ==========================================
print("4. Building Skeletons and Interpolating...")
years = list(range(2013, 2024))
all_products = pd.concat([master_mfn['ProductCode'], master_pref['ProductCode']]).dropna().unique()
pref_exporters = master_pref['iso3_exporter'].dropna().unique()

mfn_idx = pd.MultiIndex.from_product([['CHN'], years, all_products], names=['iso3_reporter', 'Year', 'ProductCode'])
mfn_skeleton = pd.DataFrame(index=mfn_idx).reset_index()

pref_idx = pd.MultiIndex.from_product([['CHN'], pref_exporters, all_products, years], names=['iso3_reporter', 'iso3_exporter', 'ProductCode', 'Year'])
pref_skeleton = pd.DataFrame(index=pref_idx).reset_index()

mfn_full = pd.merge(mfn_skeleton, master_mfn, on=['iso3_reporter', 'Year', 'ProductCode'], how='left')
pref_full = pd.merge(pref_skeleton, master_pref, on=['iso3_reporter', 'iso3_exporter', 'ProductCode', 'Year'], how='left')

mfn_full = mfn_full.sort_values(['iso3_reporter', 'ProductCode', 'Year'])
mfn_full[['mfn', 'mfnAVE']] = mfn_full.groupby(['iso3_reporter', 'ProductCode'])[['mfn', 'mfnAVE']].ffill().bfill()

pref_full = pref_full.sort_values(['iso3_reporter', 'iso3_exporter', 'ProductCode', 'Year'])
pref_full[['pref', 'prefAVE']] = pref_full.groupby(['iso3_reporter', 'iso3_exporter', 'ProductCode'])[['pref', 'prefAVE']].ffill().bfill()

# ==========================================
# STEP 5: FINAL MERGE & CEILING RULE
# ==========================================
print("5. Applying Ceiling Rule and Exporting...")
final_panel = pd.merge(
    pref_full, 
    mfn_full[['iso3_reporter', 'Year', 'ProductCode', 'mfn', 'mfnAVE']], 
    on=['iso3_reporter', 'Year', 'ProductCode'], 
    how='left'
)

final_panel['Applied_Tariff'] = final_panel['pref'].fillna(final_panel['mfn'])
final_panel['Applied_Tariff_AVE'] = final_panel['prefAVE'].fillna(final_panel['mfnAVE'])

final_panel['Applied_Tariff'] = np.fmin(final_panel['Applied_Tariff'], final_panel['mfn'])
final_panel['Applied_Tariff_AVE'] = np.fmin(final_panel['Applied_Tariff_AVE'], final_panel['mfnAVE'])

final_panel = final_panel.dropna(subset=['Applied_Tariff', 'Applied_Tariff_AVE'], how='all')

try:
    final_panel.to_stata('CHN_Final_Applied_Tariffs_2013_2023.dta', write_index=False, version=118)
    print("\n--- DONE! CHINA DATASET IS COMPLETE! ---")
    print(f"File successfully saved: CHN_Final_Applied_Tariffs_2013_2023.dta")
except Exception as e:
    print(f"\nError saving to Stata: {e}")
    final_panel.to_csv('CHN_Final_Applied_Tariffs_2013_2023.csv', index=False)
    print("Saved as CSV instead due to Stata export error.")
```

```python
import os
import pandas as pd
import numpy as np

print("--- STARTING ULTIMATE INDIA MASTER SCRIPT ---")

# ==========================================
# STEP 1: REFERENCE MAPS
# ==========================================
print("1. Loading reference maps...")
cl = pd.read_csv('country_list.csv', sep=';', encoding='latin-1')
cl.columns = [str(c).strip().replace('\ufeff', '').replace('"', '').replace("'", "") for c in cl.columns]
cl.columns = ['ISO3', 'CountryCode', 'CountryName']

pb = pd.read_csv('preference_beneficials.csv', sep=',', encoding='latin-1')
pb.columns = [str(c).strip().replace('\ufeff', '').replace('"', '').replace("'", "") for c in pb.columns]
pb.columns = ['RegionCode', 'RegionName', 'Partner', 'PartnerName']

iso_correction = {
    'ROM': 'ROU', 'ZAR': 'COD', 'TMP': 'TLS', 'SUD': 'SDN',
    'SAM': 'WSM', 'SER': 'SRB', 'MNE': 'MNE', 'ANT': 'CUW',
    'WLF': 'WLF', 'SVK': 'SVK'
}

cl['CountryCode'] = pd.to_numeric(cl['CountryCode'], errors='coerce')
iso_dict = cl.set_index('CountryCode')['ISO3'].to_dict()

def to_official_iso(numeric_code):
    try:
        wits_iso = iso_dict.get(int(float(numeric_code)))
        return iso_correction.get(wits_iso, wits_iso)
    except:
        return np.nan

pb['iso3_exporter'] = pb['Partner'].apply(to_official_iso)

# ==========================================
# STEP 2: LOAD AND MERGE TARIFFS
# ==========================================
print("2. Loading and cleaning yearly data for India...")
mfn_drop = ['Sum_Of_Rates', 'Min_Rate', 'Max_Rate', 'Nbr_NA_Lines', 'Nbr_Free_Lines', 
            'Nbr_AVE_Lines', 'Nbr_Dutiable_Lines', 'TotalNoOfValidLines', 'TotalNoOfLines', 'NomenCode']

pref_drop = ['ExcludedFrom', 'Sum_Of_Rates', 'Min_Rate', 'Max_Rate', 'TotalNoOfLines', 
             'Nbr_Pref_Lines', 'Nbr_MFN_Lines', 'Nbr_NA_Lines', 'NomenCode', 'EstCode']

def clean_tariff_file(file_path, rate_name, drop_list):
    if not os.path.exists(file_path):
        return pd.DataFrame() # Gracefully return empty dataframe if file is missing
    
    df = pd.read_csv(file_path, encoding='latin-1', low_memory=False)
    
    if 'SimpleAverage' in df.columns:
        df = df.rename(columns={'SimpleAverage': rate_name})
        
    df = df.drop(columns=[c for c in drop_list if c in df.columns], errors='ignore')
    
    if 'ProductCode' in df.columns:
        df['ProductCode'] = df['ProductCode'].astype(str).str.replace(".0", "", regex=False).str.zfill(6)
        
    return df

all_mfn, all_pref = [], []

for year in range(2013, 2024):
    # Adjust these paths if your India folders are named slightly differently!
    mfn_path = f"india/mfn/MFN_IND_{year}.CSV"
    mfn_ave_path = f"india/mfn ave/AVEMFN_IND_{year}.CSV"
    pref_path = f"india/pref/PREF_IND_{year}.CSV"
    pref_ave_path = f"india/pref ave/AVEPREF_IND_{year}.CSV"
    
    m = clean_tariff_file(mfn_path, 'mfn', mfn_drop)
    ma = clean_tariff_file(mfn_ave_path, 'mfnAVE', mfn_drop)
    p = clean_tariff_file(pref_path, 'pref', pref_drop)
    pa = clean_tariff_file(pref_ave_path, 'prefAVE', pref_drop)
    
    # Merge MFN (Handles missing MFN AVE files gracefully)
    if not m.empty:
        if ma.empty:
            all_mfn.append(m)
        else:
            all_mfn.append(pd.merge(m, ma, on=['Reporter_ISO_N', 'Year', 'ProductCode'], how='outer'))
            
    # Merge PREF (Handles missing PREF AVE files gracefully)
    if not p.empty:
        if pa.empty:
            pref_merge = p
        else:
            pref_merge = pd.merge(p, pa, on=['Reporter_ISO_N', 'Year', 'ProductCode', 'Partner'], how='outer')
            
        pref_bilateral = pd.merge(pref_merge, pb[['RegionCode', 'iso3_exporter']], left_on='Partner', right_on='RegionCode', how='inner')
        all_pref.append(pref_bilateral)

master_mfn = pd.concat(all_mfn, ignore_index=True) if all_mfn else pd.DataFrame()
master_pref = pd.concat(all_pref, ignore_index=True) if all_pref else pd.DataFrame()

if master_mfn.empty or master_pref.empty:
    raise ValueError("CRITICAL: Failed to load data. Please check your India folder paths.")

master_mfn['iso3_reporter'] = 'IND'
master_pref['iso3_reporter'] = 'IND'

print(f"Loaded {len(master_mfn)} raw MFN rows and {len(master_pref)} raw PREF rows.")

# ==========================================
# STEP 3: TETI SPIKE FILTER (VECTORIZED)
# ==========================================
print("3. Applying time-series spike filter...")
master_pref = master_pref.groupby(
    ['iso3_reporter', 'iso3_exporter', 'ProductCode', 'Year']
).agg({'pref': 'min', 'prefAVE': 'min'}).reset_index()

master_pref = master_pref.sort_values(['iso3_reporter', 'iso3_exporter', 'ProductCode', 'Year'])

# Safe Vectorized calculation
if 'prefAVE' in master_pref.columns:
    master_pref['diff'] = master_pref.groupby(['iso3_reporter', 'iso3_exporter', 'ProductCode'])['prefAVE'].diff()
    master_pref.loc[master_pref['diff'] > 0, ['pref', 'prefAVE']] = np.nan
    master_pref = master_pref.drop(columns=['diff'])

# ==========================================
# STEP 4: SKELETON MATRIX & INTERPOLATION
# ==========================================
print("4. Building Skeletons and Interpolating (Applying ffill & bfill)...")
years = list(range(2013, 2024))
all_products = pd.concat([master_mfn['ProductCode'], master_pref['ProductCode']]).dropna().unique()
pref_exporters = master_pref['iso3_exporter'].dropna().unique()

mfn_idx = pd.MultiIndex.from_product([['IND'], years, all_products], names=['iso3_reporter', 'Year', 'ProductCode'])
mfn_skeleton = pd.DataFrame(index=mfn_idx).reset_index()

pref_idx = pd.MultiIndex.from_product([['IND'], pref_exporters, all_products, years], names=['iso3_reporter', 'iso3_exporter', 'ProductCode', 'Year'])
pref_skeleton = pd.DataFrame(index=pref_idx).reset_index()

mfn_full = pd.merge(mfn_skeleton, master_mfn, on=['iso3_reporter', 'Year', 'ProductCode'], how='left')
pref_full = pd.merge(pref_skeleton, master_pref, on=['iso3_reporter', 'iso3_exporter', 'ProductCode', 'Year'], how='left')

# Forward Fill first, then Backward Fill to catch early gaps like 2013!
mfn_full = mfn_full.sort_values(['iso3_reporter', 'ProductCode', 'Year'])
mfn_full[['mfn', 'mfnAVE']] = mfn_full.groupby(['iso3_reporter', 'ProductCode'])[['mfn', 'mfnAVE']].ffill().bfill()

pref_full = pref_full.sort_values(['iso3_reporter', 'iso3_exporter', 'ProductCode', 'Year'])
pref_full[['pref', 'prefAVE']] = pref_full.groupby(['iso3_reporter', 'iso3_exporter', 'ProductCode'])[['pref', 'prefAVE']].ffill().bfill()

# ==========================================
# STEP 5: FINAL MERGE & CEILING RULE
# ==========================================
print("5. Applying Ceiling Rule and Exporting to CSV...")
final_panel = pd.merge(
    pref_full, 
    mfn_full[['iso3_reporter', 'Year', 'ProductCode', 'mfn', 'mfnAVE']], 
    on=['iso3_reporter', 'Year', 'ProductCode'], 
    how='left'
)

# Ensure AVE columns exist even if ALL AVE files were missing for some reason
if 'prefAVE' not in final_panel.columns: final_panel['prefAVE'] = np.nan
if 'mfnAVE' not in final_panel.columns: final_panel['mfnAVE'] = np.nan

# Create Fontagné-compliant Applied Tariffs
final_panel['Applied_Tariff'] = final_panel['pref'].fillna(final_panel['mfn'])
final_panel['Applied_Tariff_AVE'] = final_panel['prefAVE'].fillna(final_panel['mfnAVE'])

# Apply Teti's Ceiling Rule mathematically
final_panel['Applied_Tariff'] = np.fmin(final_panel['Applied_Tariff'], final_panel['mfn'])
final_panel['Applied_Tariff_AVE'] = np.fmin(final_panel['Applied_Tariff_AVE'], final_panel['mfnAVE'])

# Clean up empty rows
final_panel = final_panel.dropna(subset=['Applied_Tariff', 'Applied_Tariff_AVE'], how='all')

# EXPORT
final_panel.to_csv('IND_Final_Applied_Tariffs_2013_2023.csv', index=False)

print("\n--- DONE! INDIA DATASET IS COMPLETE! ---")
print(f"File successfully saved: IND_Final_Applied_Tariffs_2013_2023.csv")
print(f"Final shape: {final_panel.shape}")
```

```python
import pandas as pd
import os

print("--- STARTING NOMENCODE RESCUE ---")

# 1. Helper to find MFN files
def find_mfn_file(iso, year):
    target = f"MFN_{iso}_{year}.CSV"
    for root, dirs, files in os.walk('.'):
        for f in files:
            if f.lower() == target.lower():
                return os.path.join(root, f)
    return None

# 2. Build a Lookup Table for NomenCodes
print("Scanning raw files for NomenCodes...")
nomen_data = []
target_importers = ['CHN', 'IND', 'PAK', 'KAZ'] 

for iso in target_importers:
    for year in range(2013, 2024):
        file_path = find_mfn_file(iso, year)
        if file_path:
            # We only need to read ONE row to get the NomenCode for the whole year!
            df_temp = pd.read_csv(file_path, encoding='latin-1', nrows=1)
            if 'NomenCode' in df_temp.columns:
                nomen = df_temp['NomenCode'].iloc[0]
                nomen_data.append({'iso3_reporter': iso, 'Year': year, 'NomenCode': nomen})
            else:
                print(f"  [!] NomenCode column missing inside {file_path}")
        else:
            print(f"  [!] Could not find MFN file for {iso} {year} to extract NomenCode.")

nomen_lookup = pd.DataFrame(nomen_data)
print("\nLookup Table Built:")
print(nomen_lookup.head())

# 3. Inject into the Master Panel
print("\nLoading Master CSV Panel (this takes a moment for 23M rows)...")
master_df = pd.read_csv('MASTER_Gravity_Panel_2013_2023.csv', low_memory=False)

print("Merging NomenCode...")
master_df = pd.merge(master_df, nomen_lookup, on=['iso3_reporter', 'Year'], how='left')

# 4. Save the corrected Master Panel
print("Saving corrected Master Panel...")
master_df.to_csv('MASTER_Gravity_Panel_2013_2023.csv', index=False)

print("\n--- RESCUE COMPLETE! NomenCode restored! ---")
print(f"Final columns: {master_df.columns.tolist()}")
```

```python
import pandas as pd
import numpy as np

print("--- STARTING HS17 TO HS12 HARMONIZATION ---")

# 1. Load the Concordance File (FIXED ENCODING)
print("Loading concordance table...")
conc = pd.read_csv('hs17to12.csv', dtype=str, encoding='latin-1')

# Clean up column names (stripping any accidental leading/trailing spaces in the CSV)
conc.columns = [str(c).strip() for c in conc.columns]

# Keep only the two columns we need and rename them for easy merging
conc = conc[['Product Code', 'HS 2012 Product Code']]
conc = conc.rename(columns={'Product Code': 'HS17', 'HS 2012 Product Code': 'HS12'})

# Ensure they are formatted as 6-digit strings
conc['HS17'] = conc['HS17'].str.zfill(6)
conc['HS12'] = conc['HS12'].str.zfill(6)

# Drop any duplicate mapping pairs in the concordance file itself
conc = conc.drop_duplicates()

# 2. Load the Master Panel
print("Loading Master Tariff Panel (this takes a moment)...")
df = pd.read_csv('MASTER_Gravity_Panel_2013_2023.csv', low_memory=False)
df['ProductCode'] = df['ProductCode'].astype(str).str.zfill(6)

# 3. Split the data by Nomenclature
# H4 is already HS12. We only need to touch H5 (HS17)
print("Splitting dataset by nomenclature...")
df_h4 = df[df['NomenCode'] == 'H4'].copy()
df_h5 = df[df['NomenCode'] == 'H5'].copy()

print(f"  Rows already in HS12 (H4): {len(df_h4)}")
print(f"  Rows needing conversion (H5): {len(df_h5)}")

# 4. Map the H5 rows to HS12
print("\nMapping HS17 codes back to HS12...")
df_h5_mapped = pd.merge(df_h5, conc, left_on='ProductCode', right_on='HS17', how='inner')

# Replace the ProductCode column with the mapped HS12 code
df_h5_mapped['ProductCode'] = df_h5_mapped['HS12']

# Drop the temporary mapping columns
df_h5_mapped = df_h5_mapped.drop(columns=['HS17', 'HS12'])

# 5. Recombine the dataset
print("Recombining dataset...")
df_harmonized = pd.concat([df_h4, df_h5_mapped], ignore_index=True)

# 6. Aggregate Many-to-One splits
print("Averaging tariffs for many-to-one HS splits (Fontagné Method)...")
# We group by the core identifiers and take the mean of the tariffs
numeric_cols = ['mfn', 'mfnAVE', 'pref', 'prefAVE', 'Applied_Tariff', 'Applied_Tariff_AVE']

df_final = df_harmonized.groupby(
    ['iso3_reporter', 'iso3_exporter', 'Year', 'ProductCode'], 
    as_index=False
)[numeric_cols].mean()

# 7. Final Export
print(f"\nExporting Harmonized CSV...")
df_final.to_csv('MASTER_Tariffs_HS12_Harmonized.csv', index=False)

print("\n--- HARMONIZATION COMPLETE! ---")
print(f"Final Harmonized Rows: {len(df_final)}")
```

```python
import pandas as pd
import numpy as np

print("--- STARTING ULTIMATE HS12 HARMONIZATION (H4 + H5 + H6) ---")

# ==========================================
# 1. LOAD CONCORDANCE FILES
# ==========================================
print("Loading concordance tables...")

def prep_concordance(file_path):
    # Read with latin-1 to avoid special character errors
    conc = pd.read_csv(file_path, dtype=str, encoding='latin-1')
    
    # Clean up column names
    conc.columns = [str(c).strip() for c in conc.columns]
    
    # Keep only target columns and rename them
    conc = conc[['Product Code', 'HS 2012 Product Code']]
    conc = conc.rename(columns={'Product Code': 'Source_Code', 'HS 2012 Product Code': 'HS12_Code'})
    
    # Format to 6-digit strings
    conc['Source_Code'] = conc['Source_Code'].str.zfill(6)
    conc['HS12_Code'] = conc['HS12_Code'].str.zfill(6)
    
    return conc.drop_duplicates()

conc_17_to_12 = prep_concordance('hs17to12.csv')
conc_22_to_12 = prep_concordance('hs22to12.csv')

# ==========================================
# 2. LOAD MASTER PANEL
# ==========================================
print("Loading Master Tariff Panel (23M rows)...")
df = pd.read_csv('MASTER_Gravity_Panel_2013_2023.csv', low_memory=False)
df['ProductCode'] = df['ProductCode'].astype(str).str.zfill(6)

# ==========================================
# 3. SPLIT BY NOMENCLATURE
# ==========================================
print("Splitting dataset by nomenclature...")
df_h4 = df[df['NomenCode'] == 'H4'].copy()
df_h5 = df[df['NomenCode'] == 'H5'].copy()
df_h6 = df[df['NomenCode'] == 'H6'].copy()

print(f"  Rows already in HS12 (H4): {len(df_h4)}")
print(f"  Rows needing HS17->HS12 (H5): {len(df_h5)}")
print(f"  Rows needing HS22->HS12 (H6): {len(df_h6)}")

# ==========================================
# 4. MAP TO HS12
# ==========================================
print("\nMapping H5 (HS17) codes back to HS12...")
df_h5_mapped = pd.merge(df_h5, conc_17_to_12, left_on='ProductCode', right_on='Source_Code', how='inner')
df_h5_mapped['ProductCode'] = df_h5_mapped['HS12_Code']
df_h5_mapped = df_h5_mapped.drop(columns=['Source_Code', 'HS12_Code'])

print("Mapping H6 (HS22) codes back to HS12...")
df_h6_mapped = pd.merge(df_h6, conc_22_to_12, left_on='ProductCode', right_on='Source_Code', how='inner')
df_h6_mapped['ProductCode'] = df_h6_mapped['HS12_Code']
df_h6_mapped = df_h6_mapped.drop(columns=['Source_Code', 'HS12_Code'])

# ==========================================
# 5. RECOMBINE AND AGGREGATE
# ==========================================
print("\nRecombining dataset...")
df_harmonized = pd.concat([df_h4, df_h5_mapped, df_h6_mapped], ignore_index=True)

print("Averaging tariffs for many-to-one HS splits (Fontagné Method)...")
numeric_cols = ['mfn', 'mfnAVE', 'pref', 'prefAVE', 'Applied_Tariff', 'Applied_Tariff_AVE']

# Group by the core bilateral-product-year identifiers and calculate the mean
df_final = df_harmonized.groupby(
    ['iso3_reporter', 'iso3_exporter', 'Year', 'ProductCode'], 
    as_index=False
)[numeric_cols].mean()

# ==========================================
# 6. FINAL EXPORT
# ==========================================
print("\nExporting Final Harmonized CSV...")
df_final.to_csv('MASTER_Tariffs_HS12_Harmonized_FULL.csv', index=False)

print("\n--- HARMONIZATION COMPLETE! ---")
print(f"Final Harmonized Rows: {len(df_final)}")
```

```python
import pandas as pd
import numpy as np
import os

print("--- STARTING GLOBAL BACI INTEGRATION (2013-2023) ---")

# 1. LOAD CONCORDANCE & TARIFFS
print("Loading master lists and tariff panel...")
country_master = pd.read_csv('BACI_country_codes.csv')
iso_map = country_master.set_index('country_code')['country_iso3'].to_dict()

tariffs = pd.read_csv('MASTER_Tariffs_HS12_Harmonized_FULL.csv', low_memory=False)
tariffs['ProductCode'] = tariffs['ProductCode'].astype(str).str.zfill(6)
tariffs['Year'] = tariffs['Year'].astype(int)

# 2. DEFINE BACI FOLDER AND YEARS
baci_folder = 'BACI'
target_importers = ['CHN', 'IND', 'PAK', 'KAZ']
all_trade_years = []

# 3. LOOP THROUGH BACI FILES
print("Processing yearly BACI files...")
for year in range(2013, 2024):
    # Naming convention based on your 2013 file
    file_name = f"BACI_HS12_Y{year}_V202601.csv"
    path = os.path.join(baci_folder, file_name)
    
    if os.path.exists(path):
        print(f" -> Reading {file_name}...")
        trade = pd.read_csv(path)
        
        # Map columns and translate ISOs
        trade['Year'] = trade['t'].astype(int)
        trade['iso3_exporter'] = trade['i'].map(iso_map)
        trade['iso3_reporter'] = trade['j'].map(iso_map)
        trade['ProductCode'] = trade['k'].astype(str).str.zfill(6)
        
        # Filter for your importers before merging to save RAM
        trade = trade[trade['iso3_reporter'].isin(target_importers)].copy()
        
        # Calculate Unit Value
        trade['TradeValue_1000USD'] = trade['v']
        trade['Quantity'] = trade['q']
        trade['Unit_Value'] = np.where(trade['Quantity'] > 0, 
                                       trade['TradeValue_1000USD'] / trade['Quantity'], 
                                       np.nan)
        
        all_trade_years.append(trade[['Year', 'iso3_exporter', 'iso3_reporter', 'ProductCode', 
                                      'TradeValue_1000USD', 'Quantity', 'Unit_Value']])
    else:
        print(f" [!] Skipping {year}: File not found.")

# Combine all trade data
master_trade = pd.concat(all_trade_years, ignore_index=True)

# 4. FINAL MERGE (Preserving all Tariff Columns)
print("\nMerging with Tariff Panel (Preserving all policy columns)...")
final_df = pd.merge(
    tariffs, 
    master_trade, 
    on=['Year', 'iso3_reporter', 'iso3_exporter', 'ProductCode'], 
    how='left'
)

# 5. PPML ZERO-FILL & TRANSFORMATIONS
print("Finalizing variables...")
final_df['TradeValue_1000USD'] = final_df['TradeValue_1000USD'].fillna(0)
final_df['Quantity'] = final_df['Quantity'].fillna(0)
final_df['ln_tariff'] = np.log(1 + (final_df['Applied_Tariff_AVE'] / 100))

# 6. EXPORT
print("Exporting final datasets...")
final_df.to_csv('ULTIMATE_Gravity_Panel_HS12_Full.csv', index=False)
final_df.to_stata('ULTIMATE_Gravity_Panel_HS12_Full.dta', write_index=False, version=118)

print("\n--- DONE! ---")
print(f"Final Row Count: {len(final_df)}")
```

```python
import pandas as pd
import numpy as np

# 1. Load the final harmonized tariff panel
# This is the version with 20,941,269 rows
input_csv = 'MASTER_Tariffs_HS12_Harmonized_FULL.csv'
print(f"Loading {input_csv}...")
df = pd.read_csv(input_csv, low_memory=False)

# 2. Optimize Data Types for Stata
print("Optimizing data types to save memory and ensure Stata compatibility...")

# Convert identifiers to 'category' to shrink file size significantly
# This also bypasses common Pandas-to-Stata string length issues
for col in ['iso3_reporter', 'iso3_exporter', 'ProductCode']:
    if col in df.columns:
        df[col] = df[col].astype(str).astype('category')

# Downcast floats to float32 (reduces memory usage by 50% vs float64)
float_cols = ['mfn', 'mfnAVE', 'pref', 'prefAVE', 'Applied_Tariff', 'Applied_Tariff_AVE']
for col in float_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').astype('float32')

# Downcast Year to int16
if 'Year' in df.columns:
    df['Year'] = df['Year'].astype('int16')

# 3. Export to Stata .dta
output_dta = 'MASTER_Tariffs_HS12_Harmonized_FULL.dta'
print(f"Exporting to {output_dta}... This may take a few minutes.")

try:
    # version 118 is the standard for modern Stata (Stata 14+)
    df.to_stata(output_dta, write_index=False, version=118)
    print("\n--- SUCCESS! ---")
    print(f"File saved as: {output_dta}")
except Exception as e:
    print(f"\nStandard export failed: {e}")
    print("Attempting export with pyreadstat...")
    import pyreadstat
    pyreadstat.write_dta(df, output_dta)
    print("Saved successfully using pyreadstat.")
```

```python
import pandas as pd
import numpy as np
import pyreadstat

print("--- FIXING CEPII GRAVITY MERGE (ROBUST STATIC EXTRACTION) ---")

# 1. Load your ultimate trade panel
print("Loading ultimate trade panel...")
final_df = pd.read_csv('ULTIMATE_Gravity_Panel_HS12_Full.csv', low_memory=False)

# 2. Extract Static Geography Properly
print("Extracting clean static geography...")
static_vars = ['iso3_o', 'iso3_d', 'dist', 'contig', 'comlang_off', 'comcol']
gravity_static = df_gravity[static_vars].copy()

# CRITICAL FIX: Drop rows where distance is blank BEFORE dropping duplicates
gravity_static = gravity_static.dropna(subset=['dist'])

# Now, keep one clean, fully populated row per country pair
gravity_static = gravity_static.drop_duplicates(subset=['iso3_o', 'iso3_d'])

# Merge Static Variables
final_df = pd.merge(
    final_df, 
    gravity_static, 
    left_on=['iso3_exporter', 'iso3_reporter'], 
    right_on=['iso3_o', 'iso3_d'], 
    how='left'
)
final_df = final_df.drop(columns=['iso3_o', 'iso3_d'])

# 3. Extract and Merge Time-Varying RTAs
print("Processing time-varying RTAs...")
rta_vars = ['year', 'iso3_o', 'iso3_d', 'fta_wto']
gravity_rta = df_gravity[rta_vars].copy()
gravity_rta['year'] = gravity_rta['year'].astype(int)

final_df = pd.merge(
    final_df, 
    gravity_rta, 
    left_on=['Year', 'iso3_exporter', 'iso3_reporter'], 
    right_on=['year', 'iso3_o', 'iso3_d'], 
    how='left'
)
final_df = final_df.drop(columns=['year', 'iso3_o', 'iso3_d'])
final_df = final_df.rename(columns={'fta_wto': 'rta'})

# 4. Forward-Fill RTAs
print("Forward-filling missing RTAs...")
final_df = final_df.sort_values(['iso3_exporter', 'iso3_reporter', 'ProductCode', 'Year'])
final_df['rta'] = final_df.groupby(['iso3_exporter', 'iso3_reporter'])['rta'].ffill()

# Fill dummy variable NAs with 0
dummy_cols = ['rta', 'contig', 'comlang_off', 'comcol']
for col in dummy_cols:
    final_df[col] = final_df[col].fillna(0)

# 5. VERIFICATION
missing_dist = final_df['dist'].isnull().sum()
print(f"\n--- VERIFICATION ---")
print(f"Total Rows: {len(final_df)}")
print(f"Rows missing distance: {missing_dist} ({(missing_dist/len(final_df))*100:.2f}%)")

if missing_dist > 0:
    missing_exporters = final_df[final_df['dist'].isnull()]['iso3_exporter'].unique()
    print(f"\nExporters failing to match in CEPII Gravity: {list(missing_exporters)[:15]}")
    print("(These are usually synthetic codes like 'Free Zones' or tiny island territories)")

# 6. Final Export
print("\nOptimizing and exporting to Stata...")
for col in ['iso3_reporter', 'iso3_exporter', 'ProductCode']:
    final_df[col] = final_df[col].astype(str)

float_cols = ['Applied_Tariff_AVE', 'TradeValue_1000USD', 'dist']
for col in float_cols:
    if col in final_df.columns:
        final_df[col] = pd.to_numeric(final_df[col], errors='coerce').astype('float32')

for col in dummy_cols:
    final_df[col] = final_df[col].astype('int8')

pyreadstat.write_dta(final_df, 'ULTIMATE_Gravity_Dataset_With_Controls.dta', version=14)
print("Saved successfully!")
```

```python
import pandas as pd

# Load data
df = pd.read_excel("res.xlsx")

# Ensure HS6 is string and correctly formatted
df["HS6"] = df["HS6"].astype(str).str.zfill(6)

# Extract HS2
df["HS2"] = df["HS6"].str[:2].astype(int)

# Map to HS sections
def map_section(hs2):
    if 1 <= hs2 <= 5: return 1
    elif 6 <= hs2 <= 14: return 2
    elif hs2 == 15: return 3
    elif 16 <= hs2 <= 24: return 4
    elif 25 <= hs2 <= 27: return 5
    elif 28 <= hs2 <= 38: return 6
    elif 39 <= hs2 <= 40: return 7
    elif 41 <= hs2 <= 43: return 8
    elif 44 <= hs2 <= 46: return 9
    elif 47 <= hs2 <= 49: return 10
    elif 50 <= hs2 <= 63: return 11
    elif 64 <= hs2 <= 67: return 12
    elif 68 <= hs2 <= 70: return 13
    elif hs2 == 71: return 14
    elif 72 <= hs2 <= 83: return 15
    elif 84 <= hs2 <= 85: return 16
    elif 86 <= hs2 <= 89: return 17
    elif 90 <= hs2 <= 92: return 18
    elif hs2 == 93: return 19
    elif 94 <= hs2 <= 96: return 20
    elif hs2 == 97: return 21

df["section"] = df["HS2"].apply(map_section)

# Aggregate
result = df.groupby("section").agg(
    avg_elasticity=("epsilon", "mean"),
    std_dev=("epsilon", "std"),
    min_elasticity=("epsilon", "min"),
    num_regressions=("epsilon", "count")
).reset_index()

# Save results to Excel
result.to_excel("HS_section_results.xlsx", index=False)

print("Saved to HS_section_results.xlsx")
```
